```python
import cv2
import matplotlib.pyplot as plt

image_path = 'input.jpeg'

gray_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10, 6))
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
image = cv2.imread('input.jpeg')
dark_image = cv2.convertScaleAbs(image, alpha=0.8, beta=-100)

plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(dark_image, cv2.COLOR_BGR2RGB))
plt.title('Dark Image')
plt.axis('off')
plt.show()
```

```python
inverted_image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(inverted_image, cv2.COLOR_BGR2RGB))
plt.title('Inverted Image')
plt.axis('off')
plt.show()
```

```python
img_neg = cv2.bitwise_not(gray_img)

plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(img_neg, cv2.COLOR_BGR2RGB))
plt.title('Negative Image')
plt.axis('off')
plt.show()
```

```python
img_flipped = cv2.flip(image, 1)

plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(img_flipped, cv2.COLOR_BGR2RGB))
plt.title('Flipped Image')
plt.axis('off')
plt.show()
```
